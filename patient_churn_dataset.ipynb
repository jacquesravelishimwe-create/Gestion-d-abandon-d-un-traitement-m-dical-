```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('patient_churn_dataset.csv', sep = ';')
data.head()
```

```python
# analyse de forme
# objectif de l'etude : prediction d'abandon des clients pour un traitement medical
# f1_score :60% et un recall : 80%

# variable target : Churned
# nombre de lignes et de colonnes : (2000, 21)
# on a 15 variables quantitatives et 6 variables qualitatives
# gestion des valeurs manquantes : il y a pas de valeurs manquantes
print(data.shape)
print(data.dtypes.value_counts())
print(data.dtypes)
print(data.isna().sum())
```

```python
df = data.copy()
# analyse de fond
# visualisation de la target : il y a un taux eleve d'abandon(plus d'abandon que de non_abandon)
# suppresion des colonnes inutiles
df = df.drop(['Last_Interaction_Date', 'PatientID'], axis = 1)
df['Churned'].value_counts(normalize = True).plot.pie()
```

```python
# signification des variables qualitatives
for col in df.select_dtypes(exclude = ['float64', 'int64']):
    plt.figure()
    df[col].value_counts().plot.pie()
    plt.show()
```

```python
# signification des variables quantitatives
# la plupart des variables ont subi une transformation
for col in df.select_dtypes(exclude = 'object'):
    plt.figure()
    sns.distplot(df[col])
    plt.show()
```

```python
# relation target/variables
# creation des categories
df1 = df[df['Churned'] == 1]
df0 = df[df['Churned'] == 0]
```

```python
# relation target/variables quantitatives
# variable age : les individus entre 50 et 70 ans n'abandonnent pas le traitement medical
# variable visits_last_year : il ya plus de visites qui se sont soldes par l'abandon du traitement medical
# variable missed_appointments : les personnes qui ratent le plus de rendez-vous sont ceux qui ont abandonnent le traitement medical
# variable days_since : on peut voir que plus les jours passent plus les personnent abandonnent le traitement medical
# variable overall satisfaction : ceux qui ont une satisfaction generale eleve n'ont pas encore abandonne le traitement
# variable wait time satisfaction : ici, une satisfaction elevee % au temps d'attente se manifeste chez les non_abandons
# variable provider rate : ils abandonnent du fait d'un taux faible de sympathie
# variable avg_out_of_pocket : ils n'abandonnent pas beaucoup lorsque le traitement est moins cher
# variable billing issue : ceux qui font le traitement contraceptif abandonnent beaucoup
# variable referrals_made : apres la dispension des recommandations, les patients abandonnent beaucoup
# variable distance : plus la distance est elevee plus les patients abandonnent le traitement

for col in df.select_dtypes(exclude = 'object'):
    plt.figure()
    sns.distplot(df1[col], label = 'churned')
    sns.distplot(df0[col], label = 'not churned')
    plt.legend()
    plt.show()
```

```python
# analyse plus detaillee
# relation target/variables qualitatives
# Il y a des categories avec plus d'abandons


# il n'existe aucune correlation entre ces variables 
col_quali = ['Gender', 'State', 'Specialty', 'Insurance_Type']
for col in df.select_dtypes(exclude = ['float64', 'int64']):
    plt.figure()
    sns.heatmap(pd.crosstab(df[col], df['Churned']), annot = True, fmt = 'd', cbar = True)
    plt.show()
```

```python
# test d'hypotheses
col_test = ['Age', 'Tenure_Months', 'Visits_Last_Year', 'Missed_Appointments', 'Days_Since_Last_Visit', 
            'Overall_Satisfaction', 'Wait_Time_Satisfaction', 'Staff_Satisfaction', 'Provider_Rating', 'Avg_Out_Of_Pocket_Cost',
           'Billing_Issues', 'Portal_Usage', 'Referrals_Made', 'Distance_To_Facility_Miles', 'Churned']
df1 = df[col_test]
df10 = df1[df1['Churned']==0]
df11 = df1[df1['Churned']==1]
print(df10.shape)
print(df11.shape)

balanced_df = df11.sample(df10.shape[0])
```

```python
from scipy.stats import ttest_ind
# les abandons sont dus au variables qu'on a etudie(significativement differents)
# H0 : les moyennes des variables sont egaux pour les abandons et non abandons

def ttest(col):
    alpha =  0.05
    stat, p = ttest_ind(balanced_df[col], 
                              df10[col])
    if p < alpha:
        return ' we reject H0'
    else:
        return 0
```

```python

for col in col_test:
    print(f'{col:-<50} {ttest(col)}')

    
# conclusion : 
# Les variables avec reject H0 influencent l'abandon
```

```python
# Nettoyage-train_split
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder

train_set, test_set = train_test_split(df, test_size = 0.2, random_state = 0)

def proprecessing(df):
    
    X = df.drop('Churned', axis = 1)
    Y = df['Churned']
    print(Y.value_counts())
    
    return X, Y
```

```python
X_train, Y_train = proprecessing(train_set)
X_test, Y_test = proprecessing(test_set)

# transformation des colonnes
numerical_column = make_column_selector(dtype_include = np.number)
categorical_column = make_column_selector(dtype_exclude = np.number)
    
numerical_pipeline = make_pipeline(StandardScaler())
categorical_pipeline = make_pipeline(OneHotEncoder())
    
preprocessor = make_column_transformer((numerical_pipeline, numerical_column),
                                          (categorical_pipeline, categorical_column))
    
```

```python
# modelisation
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import learning_curve
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC

def evaluation(model):
    
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    
    print(confusion_matrix(Y_test, Y_pred))
    print(classification_report(Y_test, Y_pred))
    
    
    N, train_score, val_score = learning_curve(model, X_train, Y_train,
                                              cv = 4, scoring = 'f1', 
                                               train_sizes = np.linspace(0.1, 1, 10))
    plt.figure(figsize = (12, 8))
    plt.plot(N, train_score.mean(axis = 1), label = 'train_score')
    plt.plot(N, val_score.mean(axis = 1), label = 'val_score')
    plt.legend()

```

```python
Random = make_pipeline(preprocessor, RandomForestClassifier(random_state = 0))
KNN = make_pipeline(preprocessor, KNeighborsClassifier())
SVM = make_pipeline(preprocessor, SVC(random_state = 0))
AdaBoost = make_pipeline(preprocessor, AdaBoostClassifier(random_state = 0))
Logistic = make_pipeline(preprocessor, LogisticRegression())

list_of_model = {
    'Random': Random,
    'KNN' : KNN,
    'SVM' : SVM,
    'Adaboost' : AdaBoost,
    'Logistic' : Logistic
}

for name, model in list_of_model.items():
    print(name)
    evaluation(model)
```
